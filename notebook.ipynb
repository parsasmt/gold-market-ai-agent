# Gold Market AI Agent

This project implements an intelligent agent for answering questions about the gold market using:

- Real-time market data
- Recent financial news
- Large Language Models (GPT-OSS-120B)
- Prompt engineering

The agent supports both English and Persian.

```python
from market_data import get_market_data
from news_fetcher import get_gold_news, format_news_for_prompt
from prompts import build_prompt
from llm_analyzer import analyze
from agent import GoldMarketAgent
```

## Market Data
Retrieve current information about:

- Gold
- US Dollar Index (DXY)
- Silver
- Crude Oil

```python
market_data = get_market_data()

market_data
```

## News Retrieval

Fetch recent news related to:

- Gold
- Inflation
- Interest rates
- Geopolitical tensions

```python
news = get_gold_news()

news
```

```python
news_text = format_news_for_prompt(news)

print(news_text)
```

## Prompt Engineering

Combine:

- Market data
- News
- User question

into a single prompt sent to the language model.

```python
question = "Is now a good time to buy gold?"

prompt = build_prompt(
    question=question,
    market_data=market_data,
    news_text=news_text
)

print(prompt)
```

## LLM Analysis

Use GPT-OSS-120B through OpenRouter to generate the final answer.

```python
response = analyze(prompt)

print(response)
```

## Complete AI Agent

```python
agent = GoldMarketAgent()

answer = agent.run(
    "What important news affected gold today?"
)

print(answer)
```

## Example Questions

```python
questions = [
    "What is the current price of gold?",
    "What important news affected gold today?",
    "Is now a good time to buy gold?",
    "How is the outlook for gold next week?",
    "آیا الان زمان مناسبی برای خرید طلا است؟"
]

for q in questions:

    print("="*100)
    print("Question:")
    print(q)

    print("\nAnswer:")
    print(agent.run(q))

    print("="*100)
```
